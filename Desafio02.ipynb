##**Aula 2:** Estatísticas, visualização de dados e distribuições

```python
import pandas as pd
import matplotlib.pyplot as plt

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
```

---
---

###**DESAFIO 01** 
*ordenar o gráfico countplot e usar a biblioteca nos demais gráficos*

Para este desafio, criamos uma variável **cod_droga** para *filtrar a tabela* para as 5 drogas mais frequentes

```python
cod_droga = dados['droga'].value_counts().index[0:5]
```

E para plotar o gráfico com a biblioteca seaborn utilizamos uma query que irá retornar as 5 drogas mais frequentes

query: dados.query('droga in @cod_droga'). Usamos o parâmetro

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(10, 8))
ax= sns.countplot(x='droga', data=dados.query('droga in @cod_droga'), order=cod_droga)
ax.set_title('Top 5 Drogas')
plt.show
```

---
---

###**DESAFIO 02** 
*melhorar a visualização alterando o tamanho da fonte*

O comando utilizado é o **fontsize**

```python
plt.figure(figsize=(10, 8))
ax= sns.countplot(x='droga', data=dados.query('droga in @cod_droga'), order=cod_droga, palette='Set2', saturation=0.75, )
ax.set_title('Top 5 Drogas', fontsize=30)
ax.set_ylabel('Quantidade', fontsize=20)
ax.set_xlabel('Droga', fontsize=20)
ax.set_xticklabels(ax.get_xticklabels(), rotation=27)
plt.show()
```

###**DESAFIO 03** 
*plotar histograma com o seaborn*

Iremos plotar um histograma da média de genes da coluna e outro com a média dos tipos celulares com o comando **.histplot**

```python
hist1 = sns.histplot(data=dados.loc[:,'g-0':'g-771'].describe().T['mean'], bins=50, color='#836FFF')
hist1.set_title('Média dos Genes', fontsize=25, color='#FFA07A')
hist1.set_xlabel('valor', fontsize=20, color='#FFA07A')
hist1.set_ylabel('quantidade', fontsize=20, color='#FFA07A')

plt.show()
```

```python

hist1 = sns.histplot(data=dados.loc[:,'c-0':'c-99'].describe().T['mean'], bins=50, color='#836FFF')
hist1.set_title('Média das Células', fontsize=25, color='#FFA07A')
hist1.set_xlabel('valor', fontsize=20, color='#FFA07A')
hist1.set_ylabel('quantidade', fontsize=20, color='#FFA07A')

plt.show()
```

###**DESAFIO 06** 
*Fazer outras análises com o bloxplot e histograma*

Podemos avaliar as top 5  drogas em relação ao tratamento

```python
plt.figure(figsize=(10, 8))
ax= sns.countplot(x='droga', hue='tratamento', data=dados.query('droga in @cod_droga'), order=cod_droga, palette='Set2', saturation=0.75, )
ax.set_title('Top 5 Drogas', fontsize=30)
ax.set_ylabel('Quantidade', fontsize=20)
ax.set_xlabel('Droga', fontsize=20)
ax.set_xticklabels(ax.get_xticklabels(), rotation=27)
plt.show()
```

A partir desse gráfico conseguimos identificar que a droga mais utilizada na verdade se trata do controle
